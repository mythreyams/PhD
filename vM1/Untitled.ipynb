```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.interpolate import UnivariateSpline
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
# read the neuron positions data frame
file = 'Y:/V10/Registration/Outputs/Neuron_Position_and_Angles.csv'
df = pd.read_csv(file)
```

```python
df.columns
```

```python
df.dropna()
sum(abs(df['Thickness_Difference_From_Neuron_Axis'])<400) / 90
```

```python
fig = plt.figure(figsize=(8,6))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')

xs = df['Soma_Center_X']
ys = -df['Soma_Center_Y']
zs1 = abs(df['Thickness_Difference_From_Neuron_Axis'])
zs2 = df['Cortical_Thickness_As_Per_Nearest_Axis_in_Avg_Surface']
ax1.scatter(xs, ys, zs1,  alpha=0.6, edgecolors='w')
ax2.scatter(xs, ys, zs2,  alpha=0.6, edgecolors='w')

ax1.set_xlabel('Soma_Center_X')
ax1.set_ylabel('Soma_Center_Y')
ax1.set_zlabel('Thickness_Difference_From_Neuron_Axis')

ax2.set_xlabel('Soma_Center_X')
ax2.set_ylabel('Soma_Center_Y')
ax2.set_zlabel('Cortical_Thickness_As_Per_Nearest_Axis_in_Avg_Surface')

plt.show()
```
