# Create M1 cluster based on rabies density

```python
# Read the landmarks and its centroid


for exp_name in ['MG48_lhs','MG48_rhs']:
    for voxel_size in [100,200,300,400,500]:
        for dens_cutoff_threshold in [0.1,0.15,0.2]:
#             exp_name = 'MG50_lhs'
#             voxel_size = 300
#             dens_cutoff_threshold = 0.15

            l = Landmarks(landmark_path+exp_name+rabies_file_pattern)

            # clean up the surface based on bounding box of the used tissue
            sg = AmiraSpatialGraph(landmark_path+exp_name+'.am')
            print(landmark_path+exp_name+'.am')
            bounds = Surface(pts = sg.pia.edge_pt_coords).surface.GetBounds()
            print(bounds)
            cleaned_pts = l.get_landmarks_in_bounding_box(bounds)

            pts_pdata = Surface(pts=cleaned_pts)

            # Compute density volume image
            densfilt = vtk.vtkPointDensityFilter()
            #densfilt.SetDensityEstimateToRelativeRadius()
            densfilt.SetDensityEstimateToFixedRadius()
            #densfilt.SetRelativeRadius(0.25)
            densfilt.SetRadius(voxel_size*1.73/2)
            densfilt.SetDensityFormToNumberOfPoints()
            
            

            x = pts_pdata.surface.GetBounds()[1] - pts_pdata.surface.GetBounds()[0]
            y = pts_pdata.surface.GetBounds()[3] - pts_pdata.surface.GetBounds()[2]
            z = pts_pdata.surface.GetBounds()[5] - pts_pdata.surface.GetBounds()[4]

            densfilt.SetSampleDimensions(int(x/voxel_size)+1,int(y/voxel_size)+1,int(z/voxel_size)+1)
            densfilt.SetInputData(pts_pdata.surface)
            densfilt.SetModelBounds(pts_pdata.surface.GetBounds())
            densfilt.Update()
            densimage = densfilt.GetOutput()
            print(exp_name,voxel_size,dens_cutoff_threshold)
            print(densimage.GetOrigin())
            print(densimage.GetScalarRange()[0],densimage.GetScalarRange()[1])
            print(densimage.GetDimensions())
            
#             
#             imagewriter = vtk.vtkTIFFWriter()
#             imagewriter.SetInputData(densimage)
#             imagewriter.SetFileName(output_path+'dens_image.tiff')
#             imagewriter.Update()
            
            # binarize the image based on density cut off (10% of max density)
            max_dens =  densimage.GetScalarRange()[1]
            thre = vtk.vtkImageThreshold()
            thre.SetInputData(densimage)
            #thre.SetOutputScalarTypeToUnsignedInt()
            thre.SetInValue(0)
            thre.SetOutValue(255)
            thre.ThresholdByLower(max_dens*dens_cutoff_threshold)
            thre.Update()

            # do connected component filtering to find M1 image
            # the largest cluster M1 will be labelled 1 by the connectivity filter
            conn = vtk.vtkImageConnectivityFilter()
            conn.SetInputData(thre.GetOutput())
            conn.Update()
            binim = conn.GetOutput()

            
#             imagewriter = vtk.vtkTIFFWriter()
#             imagewriter.SetInputData(binim)
#             imagewriter.SetFileName(output_path+'bin_image.tiff')
#             imagewriter.Update()
#             y_max = 99999999
#             for i in range(conn.GetNumberOfExtractedRegions()):
#                 if y_max > conn.GetExtractedRegionExtents().GetComponent(i,3):
#                     y_max = conn.GetExtractedRegionExtents().GetComponent(i,3)
#                     #print(conn.GetExtractedRegionExtents(),y_max)
#                     M1_region_id = i+1
                    
            # find out what is the largest cluster by comparing region sizes
#             M1_region_id = 0
#             region_size = 0
#             # find the top 1/3 y in order to qualify that we are looking at m1 and not s1
#             y_range = bounds[3] - bounds[2]
#             y_thre = bounds[2] + 2* (y_range / 3)
#             print(y_thre)
#             for i in range(conn.GetNumberOfExtractedRegions()):
                
#                 if region_size < conn.GetExtractedRegionSizes().GetComponent(i,0):
#                     region_size = conn.GetExtractedRegionSizes().GetComponent(i,0)
#                     if conn.GetExtractedRegionExtents().GetComponent(i,3) > y_thre:
#                         M1_region_id = i+1
#             print(M1_region_id)

            # make a list of connected component index and sizes
            region_inds = range(conn.GetNumberOfExtractedRegions())
            region_sizes = []
            M1_region_id = 0
#             for i in range(conn.GetNumberOfExtractedRegions()):
#                 region_sizes.append(conn.GetExtractedRegionSizes().GetComponent(i,0))
            
            sorted_inds = sorted(region_inds,key=lambda i:conn.GetExtractedRegionSizes().GetComponent(i,0),reverse=True)
#             print(sorted_inds)
#             y_range = bounds[3] - bounds[2]
#             y_thre = bounds[3] -  (y_range / 3)
#             print(bounds[3] , bounds[2])
#             print(y_range)
            
#             print((y_thre-binim.GetOrigin()[1])/binim.GetSpacing()[1])
            
            y_thre = (densimage.GetDimensions()[1]*1/2)
            #print(y_thre)
            for ind in sorted_inds:
                #print(ind,conn.GetExtractedRegionSizes().GetComponent(ind,0),conn.GetExtractedRegionExtents().GetComponent(ind,2))
                y_center = (conn.GetExtractedRegionExtents().GetComponent(ind,2)+\
                            conn.GetExtractedRegionExtents().GetComponent(ind,3))/2
                #print(ind,conn.GetExtractedRegionSizes().GetComponent(ind,0),y_center)
                
                if (y_center) < y_thre:
                    M1_region_id = ind
                    break
            print(M1_region_id)
            # all landmarks engulfed by the image are part of the valid M1 cluster
#             i = 0
#             for sorted_ind in sorted_inds: 
                
            valid_cluster = []
            for pt in l.pts:
                x = (pt[0]-binim.GetOrigin()[0])/binim.GetSpacing()[0]
                y = (pt[1]-binim.GetOrigin()[1])/binim.GetSpacing()[1]
                z = (pt[2]-binim.GetOrigin()[2])/binim.GetSpacing()[2]
                ind = list(map(int,[x,y,z]))

                if binim.GetScalarComponentAsFloat(ind[0],ind[1],ind[2],0) == M1_region_id+1:
                    valid_cluster.append(pt)

#                 i +=1
                Landmarks(pts=valid_cluster).write_landmarks\
                (output_path+'{}_{}_{}_{}.landmarksAscii'.format(exp_name,voxel_size,dens_cutoff_threshold,max_dens))
#                 print(sorted_ind,len(valid_cluster))
            
            
    
```

# Create S1 cluster based on barrels

```python
# read barrels and sbf
for exp_name in ['MG49_lhs','MG49_rhs','MG50_lhs','MG50_rhs']:
    sg = AmiraSpatialGraph(landmark_path+exp_name+'.am')

    avg_sbf_file = '/nas1/Data_Mythreya/MotorCortexProject/V4/0_Inputs/average_barrel_field.am'
    sg_sbf = AmiraSpatialGraph(avg_sbf_file,barrel_projections_present=True)

    S1_convex_hull_name = '/nas1/Data_Mythreya/MotorCortexProject/V4/0_Inputs/old_sbf/average_barrel_field_S1_convex_hull.stl'
    S1_hull = Surface(filename=S1_convex_hull_name)

    # Register the standard sbf to these barrels
    matching_barrels = MatchBarrels(sg.barrels,sg_sbf.barrels,use_barrels_alone=True)
    ref_pts,pts = matching_barrels.get_matching_barrel_centroids()

    # align barrel centroid
    icp,txmat = Landmarks(pts=pts).align_landmarks(ref_pts)
    sg_sbf.apply_transformation(txmat)
    S1_hull.apply_icp_transform(icp)
    sg_sbf.write_spatial_graph(output_path+exp_name+'_reg_sbf.am')
    S1_hull.write_surface_mesh(output_path+exp_name+'_reg_s1_hull.vtk')

    # find rabies landmarks that are within sbf
    l = Landmarks(landmark_path+exp_name+rabies_file_pattern)
    s1_valid_pts = []
    sel = vtk.vtkSelectEnclosedPoints()
    sel.SetInputData(Surface(pts=l.pts).convert_points_to_polydata(l.pts))
    sel.SetSurfaceData(S1_hull.surface)
    #sel.SetInsideOut(True)
    sel.Update()
    arr =  sel.GetOutput().GetPointData().GetArray("SelectedPoints")
    for i in range((arr.GetNumberOfTuples())):
        if (arr.GetComponent(i,0)) == 1:
            s1_valid_pts.append(l.pts[i])

    Landmarks(pts=s1_valid_pts).write_landmarks(output_path+exp_name+'_S1_landmarks.landmarksAscii')
        
```
