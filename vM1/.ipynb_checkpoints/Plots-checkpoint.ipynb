```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.interpolate import UnivariateSpline
```

# NeuN Depth Line Profiles

```python
df = pd.read_csv(root+'_vS1_NeuN_Depth_Profile.csv')
df = df.replace(0, np.NaN)
#df_d2 = df[['NeuN_Density_MG49_lhs_C4','NeuN_Density_MG49_rhs_C4','NeuN_Density_MG50_lhs_C2','NeuN_Density_MG50_rhs_C3']] 
dens_mean = df[df.columns[2:]].mean(axis=1).values
dens_std = df[df.columns[2:]].std(axis=1).values

p = plt.subplot()
x = dens_mean#df[df.columns[2:]].mean(axis=1).values
y = df['Depth_in_um']
error = dens_std
p.plot(x,y)
plt.fill_between(x, y-error, y+error)
plt.savefig(root+'vS1_Avg_Depth_Profile.eps',format='eps')
```
