# Surface Resolutions

```python
input_path = 'Y:/V11/vM1_Ref_Frame/Outputs/Alignment_Analysis/Surface_Resolution_Selection/'
```

```python
#df_surf = pd.read_csv(input_path+'Resolution_Selection_Stats.csv')
df_surf = pd.read_csv(input_path+'Alignmen_Stats.csv')
```

```python
def get_surf_selection_error(df,res_used,present_res):
    df_align = df[df['Aligned_Using_Surf_Resolution']==res_used]
    df_surf_type = df_align[df_align['Present_Surf_Resolution']==present_res]
    pia_mean = df_surf_type['Pia_Mean_Dist_3D'].mean()
    wm_mean = df_surf_type['WM_Mean_Dist_3D'].mean()
    pia_std = df_surf_type['Pia_STD_Dist_3D'].mean()
    wm_std = df_surf_type['WM_STD_Dist_3D'].mean()
    
    return (pia_mean+wm_mean)/2,(pia_std+wm_std)/2
```

```python
fig, axs = plt.subplots(nrows=1,ncols=1, sharex=False,sharey=True,figsize=(8,8))
for res_used in [50,100,200,300,400,450,500]:
    mean,std = get_surf_selection_error(df_surf,res_used,50)
    print(mean,std)
    plt.bar(str(res_used),mean,yerr=std)
#axs.set(ylabel='Surface_Misalignment in micro meters')
#axs.set(xlabel='Voxel size of surface mesh')
```

# Global Alignemnt methods

```python
input_path ='Y:/V11/vM1_Ref_Frame/Outputs/Alignment_Analysis/Global_Alignment/'
```

```python
df_surf = pd.read_csv(input_path+'Stats.csv')
df_centers = pd.read_csv(input_path+'Stats_Centers.csv')
```

```python
#for align_type in ['Using_Local_Centers','Using_Local_Centers','Using_Local_Centers']:
def get_surf_error(df,align_type,br):
    df_align = df[df['Alignment_Type']==align_type]
    df_surf_type = df_align[df_align['Surface_Type']==br]
    surf_error_mean = (df_surf_type['Pia_Mean_Dist_3D'].mean()+df_surf_type['WM_Mean_Dist_3D'].mean())/2
    surf_error_std = (df_surf_type['Pia_STD_Dist_3D'].mean()+df_surf_type['WM_STD_Dist_3D'].mean())/2
    return surf_error_mean,surf_error_std
```

```python
#for align_type in ['Using_Local_Centers','Using_Local_Centers','Using_Local_Centers']:
def get_center_error(df,align_type,br):
    df_align = df[df['Alignment_Type']==align_type]
    df_surf_type = df_align[df_align['Surface_Type']==br]
    surf_error_mean = df_surf_type['Mean'].mean()
    surf_error_std = df_surf_type['STD'].mean()
    #surf_error_std = (df_surf_type['Pia_STD_Dist_3D'].mean()+df_surf_type['WM_STD_Dist_3D'].mean())/2
    return surf_error_mean,surf_error_std
```

```python
for align_type in ['Using_Local_Centers','Using_Local_Surfaces','Using_Surface']:
    
    wb_surf_error_mean,wb_surf_error_std = get_surf_error(df_surf,align_type,'WB')
    s1_surf_error_mean,s1_surf_error_std = get_surf_error(df_surf,align_type,'vS1')
    m1_surf_error_mean,m1_surf_error_std = get_surf_error(df_surf,align_type,'vM1')
    
    wb_center_error_mean,wb_center_error_std = get_center_error(df_centers,align_type,'WB')
    s1_center_error_mean,s1_center_error_std = get_center_error(df_centers,align_type,'vS1')
    m1_center_error_mean,m1_center_error_std = get_center_error(df_centers,align_type,'vM1')
    print('Surfs')
    print(wb_surf_error_mean,wb_surf_error_std,s1_surf_error_mean,s1_surf_error_std,m1_surf_error_mean,m1_surf_error_std)
    print('Centers')
    print(wb_center_error_mean,wb_center_error_std,s1_center_error_mean,s1_center_error_std,m1_center_error_mean,m1_center_error_std)
```

```python
#figure(num=None, figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k')
fig, axs = plt.subplots(nrows=3,ncols=2, sharex=False,sharey=True,figsize=(12,12))
i = 0
for align_type in ['Using_Local_Centers','Using_Local_Surfaces','Using_Surface']:
    
    wb_surf_error_mean,wb_surf_error_std = get_surf_error(df_surf,align_type,'WB')
    s1_surf_error_mean,s1_surf_error_std = get_surf_error(df_surf,align_type,'vS1')
    m1_surf_error_mean,m1_surf_error_std = get_surf_error(df_surf,align_type,'vM1')
    
    wb_center_error_mean,wb1_center_error_std = get_center_error(df_centers,align_type,'WB')
    s1_center_error_mean,s1_center_error_std = get_center_error(df_centers,align_type,'vS1')
    m1_center_error_mean,m1_center_error_std = get_center_error(df_centers,align_type,'vM1')
    
    axs[i][0].bar('WB_Surface_Error',wb_surf_error_mean,yerr=wb_surf_error_std,)
    axs[i][0].bar('vS1_Surface_Error',s1_surf_error_mean,yerr=s1_surf_error_mean)
    axs[i][0].bar('vM1_Surface_Error',m1_surf_error_mean,yerr=m1_surf_error_mean)
    
    axs[i][1].bar('WB_Center_Error',wb_center_error_mean,yerr=wb_surf_error_std)
    axs[i][1].bar('vS1_Center_Error',s1_center_error_mean,yerr=s1_surf_error_mean)
    axs[i][1].bar('vM1_Center_Error',m1_center_error_mean,yerr=m1_surf_error_mean)
    
    #for ax in axs.flat:
    axs[i][0].set( ylabel=align_type)
    
    i = i +1

plt.savefig(input_path+'global_alignments.eps',format='eps')
plt.savefig(input_path+'global_alignments.jpg',format='jpg')
```
