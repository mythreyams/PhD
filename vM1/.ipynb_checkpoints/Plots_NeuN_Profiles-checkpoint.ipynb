```python
def interpolate_array(old,new_length=43):
    old_indices = np.arange(0,len(old))
    #new_length = 43
    new_indices = np.linspace(0,len(old)-1,new_length)
    spl = UnivariateSpline(old_indices,old,k=2,s=0,check_finite=True)
    new_array = spl(new_indices)

    return new_array
```

```python
def get_half_max_index(profile):
    print(profile)
    half_max = (profile.min()+profile.max())/2
    print(half_max)
    for i in range(len(profile)-1):
        if profile[0] < profile[len(profile)-1]:
            if profile[i]<half_max and profile[i+1]>=half_max:
                return i+1
        else:
            if profile[i]>half_max and profile[i+1]<=half_max:
                return i
```

```python
def get_half_max_index(profile):
    print(profile)
    half_max = (profile.min()+profile.max())/2
    print(half_max)
    for i in range(len(profile)-1):
        if profile[0] < profile[len(profile)-1]:
            if profile[i]<half_max and profile[i+1]>=half_max:
                return i+1
        else:
            if profile[i]>half_max and profile[i+1]<=half_max:
                return i
```

```python
# match peaks within layers
df_layers = pd.read_csv(layers_path+'MG50_lhs_vM1_Borders.csv')
l1border = df_layers['L1_Border'].mean()
l23border = df_layers['L2_3_Border'].mean()
l5aborder = df_layers['L5A_Border'].mean()
l5bborder = df_layers['L5B_Border'].mean()
```

```python
df_layerwise_scales = pd.DataFrame()
df_scaled_profile = pd.DataFrame()
scaled_profile_array = np.zeros([52])
df_scaled_profile['Depth_in_um'] = df_2d['Depth_in_um']

l1_2d_peak = df_2d['NeuN_Density_MG50_lhs_0'][0:abs(int(l1border/50))].max()
l1_3d_peak = df_3d['NeuN_Density_MG50_lhs_0'][0:abs(int(l1border/50))].max()
df_layerwise_scales['L1_Scale'] = [l1_3d_peak/l1_2d_peak]
scaled_profile_array[0:abs(int(l1border/50))] = df_2d['NeuN_Density_MG50_lhs_0'][0:abs(int(l1border/50))]*l1_3d_peak/l1_2d_peak

l23_2d_peak = df_2d['NeuN_Density_MG50_lhs_0'][abs(int(l1border/50)):abs(int(l23border/50))].max()
l23_3d_peak = df_3d['NeuN_Density_MG50_lhs_0'][abs(int(l1border/50)):abs(int(l23border/50))].max()
df_layerwise_scales['L2_3_Scale'] = [l23_3d_peak/l23_2d_peak]
scaled_profile_array[abs(int(l1border/50)):abs(int(l23border/50))] = df_2d['NeuN_Density_MG50_lhs_0'][abs(int(l1border/50)):abs(int(l23border/50))]*l23_3d_peak/l23_2d_peak

l5a_2d_peak = df_2d['NeuN_Density_MG50_lhs_0'][abs(int(l23border/50)):abs(int(l5aborder/50))].max()
l5a_3d_peak = df_3d['NeuN_Density_MG50_lhs_0'][abs(int(l23border/50)):abs(int(l5aborder/50))].max()
df_layerwise_scales['L5A_Scale'] = [l5a_3d_peak/l5a_2d_peak]
scaled_profile_array[abs(int(l23border/50)):abs(int(l5aborder/50))] = df_2d['NeuN_Density_MG50_lhs_0'][abs(int(l23border/50)):abs(int(l5aborder/50))]*l5a_3d_peak/l5a_2d_peak

l5b_2d_peak = df_2d['NeuN_Density_MG50_lhs_0'][abs(int(l5aborder/50)):abs(int(l5bborder/50))].mean()
l5b_3d_peak = df_3d['NeuN_Density_MG50_lhs_0'][abs(int(l5aborder/50)):abs(int(l5bborder/50))].mean()
df_layerwise_scales['L5B_Scale'] = [l5b_3d_peak/l5b_2d_peak]
scaled_profile_array[abs(int(l5aborder/50)):abs(int(l5bborder/50))] = df_2d['NeuN_Density_MG50_lhs_0'][abs(int(l5aborder/50)):abs(int(l5bborder/50))]*l5b_3d_peak/l5b_2d_peak


l6_2d_peak = df_2d['NeuN_Density_MG50_lhs_0'][abs(int(l5bborder/50)):52].max()
l6_3d_peak = df_3d['NeuN_Density_MG50_lhs_0'][abs(int(l5bborder/50)):52].max()
df_layerwise_scales['L6_Scale'] = [l6_3d_peak/l6_2d_peak]
scaled_profile_array[abs(int(l5bborder/50)):52] = df_2d['NeuN_Density_MG50_lhs_0'][abs(int(l5bborder/50)):52]*l6_3d_peak/l6_2d_peak
```
